```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
# Creating the dataset from the table
data = pd.DataFrame({
    "No.": range(1, 21),
    "Label": ["Positive"] * 10 + ["Negative"] * 10,
    "Score": [25, 21, 20, 19, 18, 17, 14, 13, 10, 5, 18, 15, 13, 12, 10, 8, 5, 4, 3, 3]
})
```

```python
# Convert labels to binary format (Positive = 1, Negative = 0)
data["Label"] = data["Label"].apply(lambda x: 1 if x == "Positive" else 0)
```

```python
# Compute ROC curve
fpr, tpr, thresholds = roc_curve(data["Label"], data["Score"])
roc_auc = auc(fpr, tpr)
```

```python
# Print FPR, TPR, and corresponding thresholds
print("Thresholds | FPR | TPR")
for i in range(len(thresholds)):
    print(f"{thresholds[i]:.2f} | {fpr[i]:.2f} | {tpr[i]:.2f}")

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
